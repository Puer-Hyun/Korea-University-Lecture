```python
def champsveteran(summonerId : str, api_key = api_key) -> dict :
    '''
    summonerId를 인자로 받아서, 해당 유저의 숙련도를 반환하는 함수입니다.
    '''
    try:
        query = "https://kr.api.riotgames.com/lol/champion-mastery/v4/champion-masteries/by-summoner/" + summonerId + "?api_key=" + api_key
        r = requests.get(query)
        while r.status_code == 429:
            time.sleep(3)
            query = "https://kr.api.riotgames.com/lol/champion-mastery/v4/champion-masteries/by-summoner/" + summonerId + "?api_key=" + api_key
            r = requests.get(query)
    except:
        pass

    return r.json()
```

```python
# json.dumps 시켜준 것 다시 복구하기
import json

def is_json(value):
    try:
        json.loads(value)
        return True
    except (TypeError, ValueError):
        return False

for col in user_df.columns:
    if user_df[col].apply(is_json).all():
        user_df[col] = user_df[col].apply(json.loads)
```

```python
user_champs_df = pd.DataFrame(champsveteran(user_df.iloc[0]['summonerId']))
user_champs_df
```

위에서 보듯, championId, ChampionLevel 등등이 있고, lastPlayTime은 다시 변환시켜줘야 하는 것 같다. chestGranted 는 해당 시즌에 S+, S, S-의 토큰을 받았는지를 나타내는 지표이다.

```python
import datetime
def convert_time(lastPlayTime):
    timestamp_ms = lastPlayTime
    timestamp_sec = timestamp_ms / 1000  # 밀리초를 초로 변환
    dt = datetime.datetime.fromtimestamp(timestamp_sec)
    return dt
```

```python
user_champs_df['lastPlayTime'] = pd.to_datetime(user_champs_df['lastPlayTime'], unit='ms')
```

```python
# 'statPerks' column을 분리
statPerks_df = Z['statPerks'].apply(pd.Series)

def extract_styles(data):
    primary_style = data[0]
    sub_style = data[1]

    # primary style의 딕셔너리를 분해하여 새로운 딕셔너리 생성
    primary_style_dict = {'primary_'+key: value for key, value in primary_style.items() if key != 'selections'}

    # primary style의 'selections' 리스트를 각각의 column으로 분리
    for i, sel in enumerate(primary_style['selections']):
        primary_style_dict.update({f"primary_selection_{i+1}_{key}": value for key, value in sel.items()})

    # sub style도 동일하게 처리
    sub_style_dict = {'sub_'+key: value for key, value in sub_style.items() if key != 'selections'}
    for i, sel in enumerate(sub_style['selections']):
        sub_style_dict.update({f"sub_selection_{i+1}_{key}": value for key, value in sel.items()})

    # primary style과 sub style의 정보를 결합
    return {**primary_style_dict, **sub_style_dict}

# 'styles' column을 처리
styles_df = Z['styles'].apply(extract_styles).apply(pd.Series)

# 처리한 'styles' column을 원래의 DataFrame에 결합
Z_expanded = pd.concat([Z, statPerks_df, styles_df], axis=1)

# 원래의 'statPerks'와 'styles' column은 제거
Z_expanded = Z_expanded.drop(['statPerks', 'styles'], axis=1)
```
